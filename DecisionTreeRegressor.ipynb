input->Years of Experiance

output->Salary

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5,7,8,9]).reshape(-1,1)
y=np.array([3,3.5,4,4.2,6,6.5,7,7.2])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg=DecisionTreeRegressor(random_state=42)
dt_reg.fit(x_train,y_train)
```

```python
x_grid=np.linspace(1,8,100).reshape(-1,1)
y_pred=dt_reg.predict(x_grid)
plt.scatter(x,y,color='blue')
plt.plot(x_grid,y_pred,color='red')
plt.title("Decision Tree Regression")
plt.xlabel("Years of Experiance")
plt.ylabel("Salary")

plt.show()
```

```python
dt_deep=DecisionTreeRegressor(max_depth=None)
dt_deep.fit(x_train,y_train)
y_pred_deep=dt_deep.predict(x_grid)
plt.scatter(x,y)
plt.plot(x_grid,y_pred_deep,color='red')
plt.title("Overfitting Tree")


plt.show()
```

```python
dt_controlled=DecisionTreeRegressor(max_depth=3)
dt_controlled.fit(x_train,y_train)
y_pred_controlled=dt_controlled.predict(x_grid)
plt.scatter(x,y,color='blue')
plt.plot(x_grid,y_pred_controlled,color='green')
plt.title("Controlled Tree (max_depth=3)")

plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
y_test_pred_deep=dt_deep.predict(x_test)
y_test_pred_controlled=dt_controlled.predict(x_test)
print("Deep Tree Metrics")
print("MSE:",mean_squared_error(y_test,y_test_pred_deep))
print("R2 Score:",r2_score(y_test,y_test_pred_deep))
print("Controlled Tree Metrics")
print("MSE:",mean_squared_error(y_test,y_test_pred_controlled))
print("R2 Score:",r2_score(y_test,y_test_pred_controlled))
```

```python

from google.colab import drive
import pandas as pd

drive.mount('/content/drive')
path='/content/drive/My Drive/house_data.csv'


df=pd.read_csv(path)
```

```python
print("First 5 rows of the DataFrame:")
print(df.head())

print("\nDataFrame Info:")
df.info()

print("\nDescriptive Statistics:")
print(df.describe())
```

```python
feature_cols = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated']
target_col = 'price'

X = df[feature_cols]
y = df[target_col]

print(f"Features selected: {feature_cols}")
print(f"Target variable selected: {target_col}")
print("\nFirst 5 rows of feature DataFrame X:")
print(X.head())
print("\nFirst 5 rows of target Series y:")
print(y.head())
```

```python
initial_rows = len(df)
df_filtered = df[df['price'] > 0]

X_filtered = df_filtered[feature_cols]
y_filtered = df_filtered[target_col]

removed_rows = initial_rows - len(df_filtered)
print(f"Removed {removed_rows} rows where 'price' was 0.")
print(f"Shape of X after filtering: {X_filtered.shape}")
print(f"Shape of y after filtering: {y_filtered.shape}")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_filtered, y_filtered, test_size=0.25, random_state=42)

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg=DecisionTreeRegressor(random_state=42)
dt_reg.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt

y_pred_dt_reg = dt_reg.predict(X_test)

plt.figure(figsize=(8, 6))
plt.scatter(X_filtered,y_filtered, color='blue', alpha=0.6)
plt.plot([X_test,y_pred_dt_reg], 'r--', lw=2, label='Perfect Prediction')
plt.title("Decision Tree Regressor: Actual vs. Predicted House Prices")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.grid(True)
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor


rf_model = RandomForestRegressor(n_estimators=100, random_state=42)


rf_model.fit(X_train, y_train)

```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions on the test set
y_pred = rf_model.predict(X_test)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate R-squared (R2) score
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print(f"Random Forest Regressor Performance on Test Set:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2) Score: {r2:.2f}")
```

```python
print("good fit(linear regression)")
print("train mse:",mean_squared_error(y_train,y_train_pred))
print("test mse:",r2_score(y_train,y_train_pred))
```

```python
plt.scatter(x,y)
plt.plot(x,lin_reg.predict(x),color='red')
plt.title("Good Fit: Marks vs Study Hours")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```
